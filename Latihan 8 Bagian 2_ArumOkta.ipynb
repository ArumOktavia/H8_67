```python
df_incidents = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv')

print('Dataset downloaded and read into a pandas dataframe!')
```

```python
df_incidents.head()
```

```python
df_incidents.shape
```

```python
limit = 100
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
df_incidents.shape
```

```python
latitude = 37.77
longitude = -122.42
```

```python
sanfran_map = folium.Map(location=[latitude, longitude],
zoom_start=12)

sanfran_map
```

```python
incidents = folium.map.FeatureGroup()

for lat, Ing, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, ing],
            radius=5,
            
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
    
sanfran_map.add_child(incidents)
```

```python
incidents = folium.map.FeatureGroup()

for lat, ing, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, ing],
            radius=5,
            
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
    
    
latitudes = list(df_incidents.Y)
longitudes = list(df_incidents.X)
labels = list(df_incidents.Category)

for lat, Ing, label in zip(latitude, longitudes, labels):
    folium.Marker([lat, Ing], popup=label).add_to(sanfran_map)
    
sanfran_map.add_child(incidents)
```

```python
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

for lat, Ing, label in zip(df_incidents.Y, df_incidents.X,
    df_incidents.Category):
        folium.CircleMarker(
            [lat, Ing],
            radius=5,
            
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        ).add_to(sanfran_map)
    
    
sanfran_map
```

```python
from folium import plugins

sanfran_map = folium.Map(location = [latitude, longitude],
zoom_start = 12)


incidents = plugins.MarkerCluster().add_to(sanfran_map)

for lat, Ing, label, in zip(df_incidents.Y, df_incidents.X,
df_incidents.Category):
    folium.Marker(
        location=[lat, Ing],
        icon=None,
        popup=label,
    ).add_to(incidents)
    
sanfran_map
        
```

```python
df_can = pd.read_excel('https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true',
                                sheet_name='Canada by Citizenship',
                                skiprows=range(20),
                                skipfooter=2)

print('Data downloaded and read into a dataframe!')
```

```python
df_can.head()
```

```python
print(df_can.shape)
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
             
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True) 
             
df_can.columns = list(map(str, df_can.columns))
             
df_can['Total'] = df_can.sum(axis=1)
             
years = list(map(str, range(1980, 2014)))
print ('data dimensions:', df_can.shape)
```
